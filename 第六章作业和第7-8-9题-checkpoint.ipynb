第8题

```python
score['最终成绩'] = ((score.平时成绩 * 0.3 + score.卷面成绩 * 0.7))
```

```python
score['绩点']= round(score.最终成绩/10-5,0) 
```

```python
score['绩点']= numpy.where(score['绩点']>0,score['绩点'],0) 
```

```python
score['学分*绩点']= score.学分*score.绩点
```

```python
df1 = DataFrame(score['学分*绩点'].groupby(score['学号']).sum().reset_index())
```

```python
df2 = DataFrame(score['学分'].groupby(score['学号']).sum().reset_index())
```

```python
df1_df2 = pandas.merge(df1,df2,left_on='学号', right_on='学号')
```

```python
df1_df2['GPA'] = round(df1_df2['学分*绩点']/df1_df2['学分'],2)
```

```python
df1_df2['学号'] = df1_df2.学号.astype(str)
```

```python
score_area =pandas.merge(df1_df2,student,left_on='学号',
                       right_on='学号')
```

```python
#score_area
```

```python
gpa_Area = score_area.groupby(by=['生源地'],as_index =False)['GPA'].agg({'GPA平均值':numpy.mean})
```

```python
#gpa_Area
```

```python
mainColor = (42/256,87/256,141/256,1)
```

```python
plt.figure(figsize =(10,15))
plt.barh(gpa_Area['生源地'],gpa_Area['GPA平均值'],color = mainColor)
plt.title('各地区学生的GPA平均值分布情况')
```

```python
gpa_Area = gpa_Area.sort_values(by ='GPA平均值',ascending =True)
```
