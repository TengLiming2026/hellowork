```python
#------------------------------777 ----------------------------------------------------------
def ff(ef_ff) : 

    df_ff = DataFrame(ef_ff)
    ef = np.array(df_ff.ewm(span=7).mean())
    
    eff=[]
    for i in range(-7,0) :
        dx = (ef[i-1]/ef[i] - 1)
        #dx = (ef[i]/ef[i-1] - 1)

        ff = -dx*ef[i] 
        eff.append(ff)
        #print(i)
    return eff[-1]


data["va_ef_ff"] = data["va_e_f"].rolling(21).apply(ff,raw=True) 

#--------------------------------------------------------------------------------------------------------
#data['va_e_c'] = np.select([data["va_ef_ff"].shift(-3) > data["va_ef_ff"]],[data['va_e_f']],default = np.nan)
data['va_e_c'] = np.select([data["va_ef_ff"].shift(-3) > data["va_ef_ff"]],[data['va_e_f']],default = np.nan)


#plt.figure(figsize=(15, 5), dpi=128)
plt.figure(figsize=(30,6), dpi=256)
plt.title(str(tc))

plt.plot(data['va_e_f'], color = "black",  linewidth=0.3, linestyle="-")
plt.plot(data['va_ef_ff'].shift(7)*10, color = "red",  linewidth=0.3, linestyle="-")


fig = plt.figure(figsize=(30,6), dpi=256)
ax1 = fig.add_subplot(111)

#fig, ax1 = plt.subplots(1, 1)
# 2x

x = data.index
y1 = data['close'] 
ax1.plot(x, y1,color="blue", linewidth=0.3, linestyle="-")

ax2 = ax1.twinx()

y2 = data['va_e']*10
ax2.plot(x, y2, color="red", linewidth=0.3, linestyle="-")

#ax3 = ax1.twinx()
y3 = data['va_e_f']*10
ax2.plot(x, y3, color="green", linewidth=0.3, linestyle="-")


y4 = data['va_e_c']*10
ax2.plot(x, y4, color="red", linewidth=0.6, linestyle="-")

```
