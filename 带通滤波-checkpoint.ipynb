滤波器构造函数(仅介绍Butterworth滤波器)

scipy.signal.butter(N, Wn, btype='low', analog=False, output='ba')

输入参数：

N:滤波器的阶数

Wn：归一化截止频率。计算公式Wn=2*截止频率/采样频率。（注意：根据采样定理，采样频率要大于两倍的信号本身最大的频率，才能还原信号。截止频率一定小于信号本身最大的频率，所以Wn一定在0和1之间）。当构造带通滤波器或者带阻滤波器时，Wn为长度为2的列表。

btype : 滤波器类型{‘lowpass', ‘highpass', ‘bandpass', ‘bandstop'},

output : 输出类型{‘ba', ‘zpk', ‘sos'},

输出参数：

b，a: IIR滤波器的分子（b）和分母（a）多项式系数向量。output='ba'

z,p,k: IIR滤波器传递函数的零点、极点和系统增益. output= 'zpk'

sos: IIR滤波器的二阶截面表示。output= 'sos'

```python
def ssa(series, windowLen , pcn  ):
    seriesLen = len(series)
    K = seriesLen - windowLen + 1
    X = np.zeros((windowLen, K))
    for i in range(K):
        X[:, i] = series[i:i + windowLen]
    
    U, sigma, VT = np.linalg.svd(X, full_matrices=False)

    for i in range(VT.shape[0]):
        VT[i, :] *= sigma[i]
    A = VT

    rec = np.zeros((windowLen, seriesLen))
    for i in range(windowLen):
        for j in range(windowLen-1):
            for m in range(j+1):
                rec[i, j] += A[i, j-m] * U[m, i]
            rec[i, j] /= (j+1)
        for j in range(windowLen-1, seriesLen - windowLen + 1):
            for m in range(windowLen):
                rec[i, j] += A[i, j-m] * U[m, i]
            rec[i, j] /= windowLen
        for j in range(seriesLen - windowLen + 1, seriesLen):
            for m in range(j-seriesLen+windowLen, windowLen):
                rec[i, j] += A[i, j - m] * U[m, i]
            rec[i, j] /= (seriesLen - j)
    
    result = np.sum(rec[:pcn], axis=0)
    return result[-1]
    #return result
#--------------------------------------------------------------------------------------------------------  
pcn = 1
wl = 21
fwl = 3

#---------------------------------SSA-------------------------------------------------
data['c-ssa'] = data['close'].rolling(2*wl).apply(ssa,args = (wl,pcn),raw = True) 
data['ssa'] = data['gr'].rolling(2*fwl).apply(ssa,args = (fwl,pcn),raw = True) 

#-------------------------收盘价对时间t的线性回归预测值---------------------------------------
data['c-linearreg']=ta.LINEARREG(data.close, timeperiod=133)
data['linearreg']=ta.LINEARREG(data.ssa, timeperiod=14)

#---------------------------------时间序列预测值---------------------------------------------
data['tsf']=ta.TSF(data.ssa, timeperiod=14)

#---------------------------------------------------------------------------------------

 
#data['dtf'] =data['tsf'] - data['linearreg'].shift(1) 
#data['bre'] = np.select([data['tsf'] >0 ],[data['close']],default = np.nan)  #获取快慢SSA >0的时序
#data['bre'] = np.select([data['dtf'] > 0],[data['close']],default = np.nan)  #获取快慢SSA >0的时序


#---------------------------------------------------------------------------------------
plt.figure(figsize=(15, 5))
#plt.figure(figsize=(10,6), dpi=256)
plt.title(str(tc))

#plt.plot(data['gr'], color="black", linewidth=0.1, linestyle="-")
#plt.plot(data['ssa'], color="red", linewidth=0.2, linestyle="-")
#plt.plot(data['linearreg'], color="green", linewidth=0.3, linestyle="-")
#plt.plot(data['tsf'], color="blue", linewidth=0.3, linestyle="-")


plt.plot(data['close'], color="black", linewidth=0.3, linestyle="-")
plt.plot(data['c-linearreg'] , color="blue", linewidth=0.3, linestyle="-")
plt.plot(data['c-ssa'] , color="red", linewidth=0.3, linestyle="-")
```

```python
#---------------------------------买卖点策略-------------------------------------------------
#DataFrame.diff(periods=1, axis=0) ; periods：移动的幅度，int类型，默认值为1;
# axis：移动的方向，{0 or ‘index’, 1 or ‘columns’}，
# 如果为0或者’index’，则上下移动，如果为1或者’columns’，则左右移动。

#data['dtf'] =  data['tsf'] - data['ssa'] 
#data['dtf'] = data['leadsine'] - data['leadsine'].shift(1) 

data['ema20']  = data['close'].ewm(span=20).mean()
data['ema30']  = data['close'].ewm(span=30).mean()

data['dtf'] =  (data['leadsine'] > data['leadsine'].shift(1))
#data['dtf'] =  (data['leadsine'] > data['leadsine'].shift(1))  & (data['ema20']  > data['ema30'] )
#data['dtf'] =  (data['leadsine'] > data['leadsine'].shift(1))  & (data['c-ssa']  > data['c-ssa'].shift(1) )

data['ddx'] = data['dtf'] 
#---------------------------------回测收益-------------------------------------------------

data['res'] = np.select([data['ddx'] > 0],[data['close']],default = np.nan)  #获取快慢SSA >0的时序
data['attack'] = np.select([data['ddx'] > 0],[1],default = 0)#获取“双SSA斜率差 >0策略”的时序的触发信号

bs = np.array(data['attack'])

print("买卖信号A:"+str(bs[-1]))

data['buy'] = data['attack'].shift(1) #下移一行买入信号
#data['buy'] = data['attack']  #下移一行买入信号

bs2 = np.array(data['buy'])
print("买卖信号B:"+str(bs2[-1]))

#print(np.array(data['buy']))
data['base_profit'] = 1+0.01*data['pct_chg'].fillna(0) #计算基准每日收益
#data['base_profit'] = 0.01*data['pct_chg'].fillna(0) #计算基准每日收益

data['prod_base_profit'] = data['base_profit'].cumprod()-1 #计算基准每日收益累乘
#data['prod_base_profit'] = data['base_profit'].cumsum() #计算基准每日收益累乘

data['profit'] = np.select([data['buy'] == 1],[data['base_profit']],default = 1) #计算策略每日收益
data['profit'] = data['profit'].fillna(0)
#print(np.array(data['profit']))

data['prod_profit'] = data['profit'].cumprod()-1#计算策略每日收益累乘
#data['prod_profit'] = data['profit'].cumsum()#计算策略每日收益累乘


plt.figure(figsize=(15, 5))
plt.title(str(tc))

plt.plot(data['prod_profit'], color="red", linewidth=0.3, linestyle="-")
plt.plot(data['prod_base_profit'], color = "blue",  linewidth=0.3, linestyle="-")


print("策略收益率："+str(np.array(data['prod_profit'])[-1])) 
print("基准收益率："+str(np.array(data['prod_base_profit'])[-1]))
```
