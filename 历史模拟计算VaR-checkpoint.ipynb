### 历史模拟结合计算VaR

### 1. 关于VaR
* VaR(Value at Risk)，在险值，是一种投资风险评价指标
* 一日 5% VaR 可以理解为一天发生损失超过VaR的概率小于等于5%， 或者每20天中可能出现一天（因为发生概率小于等于5%）资产单日损失超过VaR的情况。

以50ETF基金为例：

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import tushare as ts
%matplotlib inline
```

#### Tushare直接读取期权价格数据

```python
data = ts.get_k_data('510050','2005-02-23','2018-11-09')
```

```python
data.head()
```

```python
len(data)
```

```python
data['d_return'] = data['close'].pct_change()
```

```python
data.head()
```

```python
help(np.percentile)        # 就是用来从小到大排序的分位数的；
```

```python
# 百分位数是从小到大进行排练；
VaR_5 = np.percentile(data.d_return.dropna(), 5)         # np.percentile方法不能有空值；
VaR_1 = np.percentile(data.d_return.dropna(), 1)
```

```python
VaR_1
```

```python
VaR_5
```

```python
# 字符串格式化1；
'50ETF一日95%的VaR为 {:.2f}%；99%的VaR为 {:.2f}%'.format(VaR_5*100,VaR_1*100) 
```

```python
# 字符串格式化2；用%对%进行转义；
'50ETF一日95%%的VaR为 %0.2f%%, 1%%的VaR为 %0.2f%%' % (VaR_5*100,VaR_1*100)
```

```python
plt.figure(figsize= (8,6))
data.d_return.hist(bins=50, alpha=0.6, color='steelblue')
plt.axvline(VaR_5, color='r', linewidth=1,label = 'VaR_5')
plt.axvline(VaR_1, color='y', linewidth=1,label = 'VaR_1')
plt.legend()
plt.xlabel('return')
plt.ylabel('count')
```

#### 从各大股票交易平台导出数据并读入

```python
data = pd.read_csv('data/510050.txt', encoding='gbk', sep='\t', parse_dates=True, header=1, index_col=0, 
                   usecols=[0,1,2,3,4,5], engine='python', skipfooter=2, skipinitialspace=True)
data.head()
```

```python
data.tail()
```

```python
data.columns
```

```python
# 计算日收益率
data['d_return'] = data['收盘'].pct_change()
data.head()
```

```python
VaR_5 = np.percentile(data.d_return.dropna(), 5)
VaR_1 = np.percentile(data.d_return.dropna(), 1)
f'50ETF一日5%_VaR为 {VaR_5*100 : .2f}%, 1%_VaR:为 {VaR_1*100 : .2f}%'       # 新型字符串格式化；
```

```python
plt.figure(figsize= (8,6))
data.d_return.hist(bins=50, alpha=0.6, color='steelblue')
plt.axvline(VaR_5, color='r', linewidth=1,label = 'VaR_5')
plt.axvline(VaR_1, color='y', linewidth=1,label = 'VaR_1')
plt.legend()
plt.xlabel('return')
plt.ylabel('count')
```
