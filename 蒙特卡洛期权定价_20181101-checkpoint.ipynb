## 2. 模拟标的资产价格

假设股票价格服从几何布朗运动，在风险中立的条件下可以得到：

$\large S_t = S_{t-1}e^{(r-\frac{1}{2}\sigma^2)\Delta t + \sigma\sqrt{\Delta t}z_t}$ 

其中：

$S_t$: t时刻的价格

$r$: 无风险利率(年化）

$\sigma$: 收益率标准差（年化）

$\Delta t$: 时间间隔

$z_t$: 服从正态分布的随机变量


```python
# 初始数据（以11月14日作为基准）
s_0 = 2.475              # 标的资产初始价格;
K = 2.45                 # 行权价格
r = 0.03                 # 无风险利率
sigma = 0.276            # 波动率(收益标准差)
T = 14/252               # 合约到期时间（年）
n = 100                  # 单次模拟的步数
N = 100                  # 模拟价格路径的次数
```

```python
# 定义函数，模拟单一路径，返回一个列表，列表的最后一个数值即为T期后的模拟价格；
def single_simulation(s_0, T, n, r, sigma):
    """
    Simulate a single price path. 
    Param n: number of steps. 
    Return: price path as a pandas Series
    """
    p = [s_0, ]
    for i in range(n):
        z = gauss(0,1)
        s_next = p[-1]*exp((r - 0.5*sigma**2)*(T/n) + sigma*sqrt(T/n)*z)
        p.append(s_next)
    return p
```

```python
# 模拟单一价格路径并作图,每次模拟出来的价格都是不尽相同的；
p = pd.Series(single_simulation(s_0, T, n, r, sigma))
p.plot(figsize=(10,6))
plt.show()
```

```python
# 模拟多个价格路径
for i in range(N):
    p = pd.Series(single_simulation(s_0, T, n, r, sigma))
    p.plot(linewidth=0.5, figsize=(10,6))
```

## 3. 计算期权价值

认购期权的到期收益：$max(s-K, 0)$

认沽期权的到期收益：$max(0, K-s)$

其中：

$K$ 为行权价格，

$s$ 为到期时标的资产的价格

```python
# 以认购期权为例，将模拟结果与行权价格比较可以得到期权到期的平均收益
payoff = []
for i in range(N):
    p = single_simulation(s_0, T, n, r, sigma)
    path_payoff = max(p[-1] - K, 0)
    payoff.append(path_payoff)
    
payoff    
```

计算到期后的平均收益

```python
payoff_mean = sum(payoff) / len(payoff)
payoff_mean
```

将平均到期收益按照无风险利率折现，可以得到期权现价的预测值。

```python
call_premium = payoff_mean * exp(-r * T)
call_premium
```

```python
# 定义函数，实现模特卡洛模拟期权定价
def price_option(stock_price=2.4, strike=2.4, sigma=0.27, r=0.035, T=40/252, n=100, N=5000, opt_type='call'):
    """
    param opt_type: 'call' / 'put'
    param stock_price: current stock price
    param strike: strike price
    param r: rist freee rate
    param sigma: annual volatility of stock price
    param T: time to expiry (days)
    param n: number of steps per simulation
    param N: number simulation
    return: expected premium of the option
    """
    # 模拟 N 条价格路径
    payoff = []
    for i in range(N):
        s = single_simulation(stock_price, T, n, r, sigma)[-1]
        if opt_type.lower() == 'call':
            path_payoff = max(s - strike, 0)
        else:
            path_payoff = max(strike - s, 0)
        payoff.append(path_payoff)
    payoff_mean = sum(payoff) / len(payoff)
    return payoff_mean * exp(-r*T)
```

```python
# 以11月14日收盘价数据代入期权定价公式，和11月14日真实收盘价格近似；
price_option(2.475, 2.45, 0.27, 0.03, T=14/252, n=100, N=50000)
```

```python
# 以11月14日收盘价数据代入期权定价公式，和11月14日真实收盘价格近似；
price_option(2.475, 2.5, 0.27, 0.03, T=14/252, n=100, N=5000)
```

```python
stock = 2.475
bottom_strike = 2.2
sigma = 0.27
r = 0.03
T = 14/252

opt_price = []
for i in range(16):
    strike = bottom_strike + i * 0.05
    opt_price.append((strike, price_option(stock, strike, sigma, r=r, T=T, opt_type='call'), 
                      price_option(stock, strike, sigma, r=r, T=T, opt_type='put')))

df = pd.DataFrame(opt_price, columns=['strike', 'call', 'put'])
df
```
